```python
columns_check=['age','trestbps','chol','thalach','oldpeak']
```

```python
for column in columns_check:
    mean=df[column].mean()
    std_dev=df[column].std()
    z_score=(df[column]-mean)/std_dev
```

```python
threshold = 3
data= df[(z_score > threshold) & (z_score < threshold)]
```

```python
cleaned_file_path = r"C:\Users\abhishekgh\Desktop\ExcelR\heart_cleaned.csv"
data.to_csv(cleaned_file_path, index=False)

print("Outliers removed. Cleaned data saved to:", cleaned_file_path)
```

```python
##Histogram Plot
columns_to_plot=['age','trestbps','chol','thalach','oldpeak','thal']

plt.figure(figsize=(15, 10))

for i, data in enumerate(columns_to_plot):
    plt.subplot(3, 2, i + 1)
    sns.histplot(df[data], bins=30, kde=True)  # Add kde for kernel density estimation
    plt.title(f'Distribution of {data}')
    plt.xlabel(data)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,roc_curve,roc_auc_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(featurse1,target,train_size=0.80,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Logistic Regression Model

```python
log_model=LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
log_predict=log_model.predict(x_test)
log_predict
```

```python
accuracy_score(y_test,log_predict)
```

```python
log_model.coef_
```

```python
log_model.intercept_
```

```python
log_model.predict_proba(x_test)
```

```python
print(classification_report(y_test,log_predict))
```

##ROC Curve

```python
log_pred=log_model.predict_proba(x_test)[:,1]
log_pred
```

```python
auc_score=roc_auc_score(y_test,log_pred)
auc_score
```

```python
fpr,tpr,thresould=roc_curve(y_test,log_pred)
```

```python
plt.plot(fpr,tpr,lw=2,label=f'AUC_SCORE:{auc_score:.2f}')
plt.legend(loc='best')
plt.plot([0,1],[0,1],linestyle='--',color='red')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.grid()
```

## Random Forest Model
